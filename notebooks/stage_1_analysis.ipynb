```python
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("Set2")
```

```python
RUNS = 24

data_seed_0 = pd.DataFrame()
data_seed_42 = pd.DataFrame()
data_seed_3407 = pd.DataFrame()

grid_seed_0 = pd.read_csv("../outputs/01_grid_mobilenet/seed_0/grid_results.csv")
grid_seed_42 = pd.read_csv("../outputs/01_grid_mobilenet/seed_42/grid_results.csv")
grid_seed_3407 = pd.read_csv("../outputs/01_grid_mobilenet/seed_3407/grid_results.csv")

for run in range(RUNS):
    run_seed_0 = pd.read_json(
        f"../outputs/01_grid_mobilenet/seed_0/run_{run:03d}/epoch_metrics.json"
    )

    run_seed_0["run"] = run
    data_seed_0 = pd.concat([data_seed_0, run_seed_0], ignore_index=True)

    run_seed_42 = pd.read_json(
        f"../outputs/01_grid_mobilenet/seed_42/run_{run:03d}/epoch_metrics.json"
    )
    run_seed_42["run"] = run
    data_seed_42 = pd.concat([data_seed_42, run_seed_42], ignore_index=True)

    run_seed_3407 = pd.read_json(
        f"../outputs/01_grid_mobilenet/seed_3407/run_{run:03d}/epoch_metrics.json"
    )
    run_seed_3407["run"] = run
    data_seed_3407 = pd.concat([data_seed_3407, run_seed_3407], ignore_index=True)

grid = pd.concat([grid_seed_0, grid_seed_42, grid_seed_3407], ignore_index=True)
```

```python
grid.columns
```

```python
grid.groupby(["optimizer", "batch_size", "dropout", "weight_decay"]).agg(
    {
        "test_accuracy": ["mean", "std"],
        "test_loss": ["mean", "std"],
        "best_val_accuracy": ["mean", "std"],
        "best_val_loss": ["mean", "std"],
    }
).sort_values(("test_accuracy", "mean"), ascending=False)
```
